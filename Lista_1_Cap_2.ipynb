```python
#2.20 Uma escada encostada diretamente contra uma parede cairá a menos que colocada em um
#certo ângulo menor que 90 graus. Dadas as variáveis comprimento e ângulo armazenando o
#comprimento da escada e o ângulo que ela forma com o solo enquanto encostada na parede,
#escreva uma expressão Python envolvendo comprimento e ângulo, que calcule a altura
#alcançada pela escada. Avalie a expressão para estes valores de comprimento e ângulo:
#(a)16 pés e 75 graus
#(b)20 pés e 0 graus
#(c)24 pés e 45 graus
#(d)24 pés e 80 graus
#Nota: Você precisará usar a fórmula trigonométrica:

#comprimento = altura / sen(angulo)

#A função sin() do módulo math toma sua entrada em radianos. Assim, você precisará
#converter o ângulo dado em graus para o ângulo dado em radianos, usando: 

#radianos = pi*graus/180

import math

def altura_escada(comprimento, angulo):
    radianos = math.pi * angulo / 180
    altura = comprimento / math.sin(radianos)
    return altura

a)
altura_a = altura_escada(16, 75)
print("Altura alcançada (a):", altura_a, "pés")

b)
altura_b = altura_escada(20, 0)
print("Altura alcançada (b):", altura_b, "pés")

c)
altura_c = altura_escada(24, 45)
print("Altura alcançada (c):", altura_c, "pés")

d)
altura_d = altura_escada(24, 80)
print("Altura alcançada (d):", altura_d, "pés")
```

```python
#2.28 Usando a técnica do Exercício 2.26, escreva instruções Python que desenham um pentágono
#com 100 pixels de comprimento de lado usando Turtle graphics. Depois faça um hexágono, um
#heptágono e um octógono.

import turtle

# Cria a tela
s = turtle.Screen()

t = turtle.Turtle()

# Função para desenhar um polígono com um determinado comprimento de lado e número de lados
def desenhar_poligono(lados, comprimento):
    angulo = 360 / lados
    for i in range(lados):
        t.forward(comprimento)
        t.left(angulo)

# Desenha um pentágono com 100 pixels de lado
desenhar_poligono(5, 100)

t.penup()
t.forward(150)
t.pendown()

# Desenha um hexágono com 100 pixels de lado
desenhar_poligono(6, 100)

t.penup()
t.forward(150)
t.pendown()

# Desenha um heptágono com 100 pixels de lado
desenhar_poligono(7, 100)

t.penup()
t.forward(150)
t.pendown()

# Desenha um octógono com 100 pixels de lado
desenhar_poligono(8, 100)

# Fecha a tela quando termina
s.bye()
```

```python
#2.34 Usando Turtle graphics, desenhe uma imagem mostrando os seis lados de um dado. Você
#pode representar cada lado dentro de um quadrado separado.
import turtle

# Cria a tela
s = turtle.Screen()

t = turtle.Turtle()

# Função para desenhar um quadrado
def draw_square():
    for _ in range(4):
        t.forward(100)
        t.right(90)

# Loop para desenhar os seis lados do dado
for i in range(6):
    # Desenha um quadrado
    draw_square()

    t.penup()
    t.forward(150)
    t.pendown()

    t.right(90)

# Fecha a tela ao finalizar
s.bye()
```

```python
#2.35 Usando Turtle graphics, desenhe as linhas de uma quadra de basquete. Você pode escolher
#as especificações da National Basketball Association (NBA) ou da International Basketball
#Federation (FIBA), que você pode encontrar facilmente na Web. 
import turtle

# Cria a tela de desenho
s = turtle.Screen()
s.title("Quadra de Basquete (NBA)")
s.bgcolor("lightgray")

t = turtle.Turtle()
t.speed(0)

# Desenha a linha de fundo
t.penup()
t.goto(-250, -180)
t.pendown()
t.forward(500)
t.left(90)
t.forward(420)
t.left(90)
t.forward(500)
t.left(90)
t.forward(420)
t.penup()
t.goto(0, 0)
t.pendown()

# Desenha o círculo central
t.circle(60)

# Desenha o círculo do lance livre
t.penup()
t.goto(0, 180)
t.pendown()
t.circle(60)

# Desenha o retângulo do garrafão
t.penup()
t.goto(-120, 180)
t.pendown()
t.forward(240)
t.left(90)
t.forward(120)
t.left(90)
t.forward(240)
t.left(90)
t.forward(120)

# Desenha as linhas de 3 pontos
t.penup()
t.goto(-250, -60)
t.pendown()
t.forward(500)
t.penup()
t.goto(-250, 60)
t.pendown()
t.forward(500)

# Desenha as linhas laterais
t.penup()
t.goto(-250, -180)
t.pendown()
t.forward(500)
t.penup()
t.goto(-250, 180)
t.pendown()
t.forward(500)

t.hideturtle()
s.mainloop()
```

```python
#2.36 Usando Turtle graphics, desenhe uma imagem mostrando as fases (visíveis) da lua conforme
#vista do seu hemisfério: quarto crescente, lua cheia, quarto minguante, lua nova. Você pode achar
#ilustrações das fases da lua na Web

import turtle

# Configuração inicial
t = turtle.Turtle()
turtle.bgcolor("black")
t.speed(0)
t.color("white")

# Desenhar a fase de quarto crescente
t.penup()
t.goto(-150, 0)
t.pendown()
t.begin_fill()
t.circle(60, 180)
t.left(90)
t.forward(120)
t.left(90)
t.circle(60, 180)
t.end_fill()

# Desenhar a fase de lua cheia
t.penup()
t.goto(0, 0)
t.pendown()
t.begin_fill()
t.circle(60)
t.end_fill()

# Desenhar a fase de quarto minguante
t.penup()
t.goto(150, 0)
t.pendown()
t.begin_fill()
t.circle(60, -180)
t.left(90)
t.forward(120)
t.left(90)
t.circle(60, -180)
t.end_fill()

# Desenhar a fase de lua nova
t.penup()
t.goto(300, 0)
t.pendown()
t.begin_fill()
t.circle(60)
t.end_fill()

t.hideturtle()

# Manter a janela aberta
turtle.done()
```
